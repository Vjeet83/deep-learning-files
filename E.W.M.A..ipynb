```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
temp_list = []
start_temp = 0

for i in range (114):
    x =np.random.randint(0,6,1)
#     print("x value " ,x)
    curr_tem = i*0.1 + x
    temp_list.append(curr_tem)
#  curr_temp = 
# type(temp_list)
temp_list
```

```python
plt.figure(figsize=(12, 6))
plt.plot(temp_list, label='Temperature Data')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.title('Synthetic Temperature Data for 114 Days')
plt.legend()
plt.grid(True)
plt.show()
```
